중고차 가격 예측

model(객체형)
year(정수형)
price(정수형)
transmission(객체형)
mileage(정수형)
fuelType(객체형)
tax(정수형)
mpg(상수형)
engineSize(상수형) 등 총 8개의 변수로 구성되어 있으며, price를 종속변수로 나머지 변수를 독립변수로 설정하여 분석을 진행함.

```python
import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/데이터 셋/P220504-01.csv')

print(train.shape)
print(train.head())
print(train.info())
print(train.describe())
print(train.isnull().sum())
```

```python
print(train['year'].value_counts())
print(train['transmission'].value_counts())
print(train['fuelType'].value_counts())
```

```python
test = pd.read_csv('/content/drive/MyDrive/데이터 셋/P220504-02.csv')
print(test.shape)
print(test.head())
print(test.info())
print(test.describe())
print(test.isnull().sum())
```

```python
train[['year','price']].corr(method='spearman')
```

```python
print(test['year'].value_counts())
print(test['transmission'].value_counts())
print(test['fuelType'].value_counts())
```

```python
test['fuelType'] = test['fuelType'].replace('Electric', 'Other')

print(test.fuelType.value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
number = ['mileage', 'tax', 'mpg', 'engineSize']
train[number] = mms.fit_transform(train[number])
test[number] = mms.transform(test[number])

ohe = OneHotEncoder(sparse=False)
ohe_train = ohe.fit_transform(train[['transmission','fuelType','year','model']])
ohe_df_train = pd.DataFrame(ohe_train, columns=ohe.get_feature_names_out())
train_df = pd.concat([ohe_df_train, train.drop(['transmission','fuelType','year','model'], axis=1)],axis=1)

ohe_test = ohe.fit_transform(test[['transmission','fuelType', 'year','model']])
ohe_df_test = pd.DataFrame(ohe_test, columns=ohe.get_feature_names_out())
test_df = pd.concat([ohe_df_test, test.drop(['transmission','fuelType','year','model'], axis=1)],axis=1)

print(train_df)
print(test_df)
```

```python
features = list(set(train_df.columns.intersection(test_df.columns)))

train = train_df[features]
test = test_df[features]

y = train_df['price']
```

```python
print(train.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import GridSearchCV

x_train, x_valid, y_train, y_valid = train_test_split(train, y,test_size=0.2,random_state=1234)

model = RandomForestRegressor(n_estimators=100, max_depth=20)
model.fit(x_train, y_train)
```

```python
params = {'n_estimators':[10, 20, 30, 40 ,50], 'max_depth':[10,20,30,40,50]}

model = RandomForestRegressor()
gscv = GridSearchCV(estimator= model, param_grid = params)
gscv.fit(x_train, y_train)
print(gscv.best_params_)
```

```python
model = RandomForestRegressor(n_estimators=50, max_depth=30)
model.fit(x_train, y_train)
print('결정계수 값:', model.score(x_train, y_train))

pred = model.predict(x_valid)
print(pred)

print('MSE 값:', mean_squared_error(y_valid, pred, squared=False))
rmse = np.sqrt(mean_squared_error(y_valid, pred, squared=False))
print('RMSE 값:', rmse)
```

모델은 약 0.99의 결정계수 값을 가지며, 독립변수들이 종속변수(price)에 변동을 약99% 설명하고 있다고 볼 수 있다.

mse(평균제곱오차)값은 2787, rmse(평균 제곱근 오차)값은 52으로 비교적 정확한 모델을 구축하였다고 생각됩니다.


```python
features = list(set(train_df.columns.intersection(test_df.columns)))

test = test_df[features]

pred = model.predict(test)

result = pd.DataFrame({'pred':pred})
print(result)
```
